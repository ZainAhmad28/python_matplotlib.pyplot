```python
import matplotlib.pyplot as plt
import numpy as np
x= np.array([40,180])
y = np.array([200,390])

font1 = {'family':'serif', 'color':'blue','size':20}
font2 = {'family':'serif', 'color':'red','size':10}
plt.plot(x,y)
plt.ylabel("ruppees")
plt.xlabel("sale", fontdict=font2)
plt.title("shop sale", fontdict=font1)
plt.grid()
plt.show()
```

```python
x= np.array([40,180])
y = np.array([200,390])

font1 = {'family':'serif', 'color':'blue','size':20}
font2 = {'family':'serif', 'color':'red','size':20}
plt.plot(x,y)
plt.ylabel("ruppees")
plt.xlabel("sale", fontdict=font2)
plt.title("shop sale", fontdict=font1)
plt.grid(axis='x')
plt.show()
```

```python
x= np.array([40,180])
y = np.array([200,390])

font1 = {'family':'serif', 'color':'blue','size':20}
font2 = {'family':'serif', 'color':'red','size':20}
plt.plot(x,y)
plt.ylabel("ruppees")
plt.xlabel("sale", fontdict=font2)
plt.title("shop sale", fontdict=font1)
plt.grid(color='yellow',ls="--",linewidth=0.5 )
plt.show()
```

```python
# labelling

import matplotlib.pyplot as plt
import numpy as np
x = np.array([20,25,45,10])
mylabel = ["cars","bikes","cycles","trucks"]

plt.pie(x, labels=mylabel)
plt.show

```

```python
# start angle parameter

import matplotlib.pyplot as plt
import numpy as np
x = np.array([20,25,45,10])
mylabel = ["cars","bikes","cycles","trucks"]


plt.pie(x, labels=mylabel , startangle = 90)
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([20,25,45,10])
mylabel = ["cars","bikes","cycles","trucks"]

plt.pie(x, labels=mylabel, startangle=30)
plt.show
```

```python
# explode the piechart

import matplotlib.pyplot as plt
import numpy as np
x = np.array([20,25,45,10])
mylabel = ["cars","bikes","cycles","trucks"]
myexplode= [0.2,0,0,0]

plt.pie(x, labels=mylabel , startangle=90 , explode=myexplode)
plt.show
```

```python
#shadow
import matplotlib.pyplot as plt
import numpy as np
x = np.array([20,25,45,10])
mylabel = ["cars","bikes","cycles","trucks"]
myexplode= [0.2,0,0,0]

plt.pie(x, labels=mylabel , startangle=90 , explode=myexplode ,shadow="true")
plt.show
```

```python
# sepcify the color

import matplotlib.pyplot as plt
import numpy as np
x = np.array([20,25,45,10])
mylabel = ["cars","bikes","cycles","trucks"]

mycolor =["black","hotpink","green","b"]

plt.pie(x, labels=mylabel , startangle=90, colors=mycolor )
plt.show
```
